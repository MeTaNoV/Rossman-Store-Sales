The given store Nr. 622 is the only one with a NaN, and when looking at the other store, they are all opened. So, let's make the assumption that the store is opened on that day.

```python
test.loc[np.isnan(test['Open']), 'Open'] = 1
test['Open'] = test['Open'].astype(int)
```

I will do a little bit of feature engineering as follows:
- refactor the date to add a `Day`, `Month` and `Year` column and drop the `Date` column.
- add a new column `isCompetition` out of `CompetitionOpenSinceMonth` and `CompetitionOpenSinceYear` and drop the latest. Also, set 0 all `CompetitionDistance` that are NaN.
- change column `Promo2` based on all `Promo2xxx` related field and drop those. Since this kind of promotion is constantly running because it is renewed every 3 months, only a proper time series model would probably capture its contribution.
This way, most of the NaN will be removed, apart from one that will be discussed in a moment.

```python
train['Day'] = train['Date'].dt.day
train['Week'] = train['Date'].dt.week
train['Month'] = train['Date'].dt.month
train['Year'] = train['Date'].dt.year

test['Day'] = test['Date'].dt.day
test['Week'] = test['Date'].dt.week
test['Month'] = test['Date'].dt.month
test['Year'] = test['Date'].dt.year

def isCompetition(row):
    if row['Year'] >= row['CompetitionOpenSinceYear'] and \
       row['Month'] >= row['CompetitionOpenSinceMonth']:
        return 1
    else:
        return 0

train['isCompetition'] = train.apply(lambda row: isCompetition(row), axis=1)
test['isCompetition'] = test.apply(lambda row: isCompetition(row), axis=1)
train.loc[np.isnan(train['CompetitionDistance']), 'CompetitionDistance'] = 0
test.loc[np.isnan(test['CompetitionDistance']), 'CompetitionDistance'] = 0

def isPromo2(row):
    if row['Promo2'] == 1 and \
       row['Year'] >= row['Promo2SinceYear'] and \
       row['Week'] >= row['Promo2SinceWeek']:
        return 1
    else:
        return 0
        
train['Promo2'] = train.apply(lambda row: isPromo2(row), axis=1)
test['Promo2'] = test.apply(lambda row: isPromo2(row), axis=1)
```

```python
drop_columns = ['Date', 'Week', \
                'CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear', \
                'Promo2SinceWeek', 'Promo2SinceYear', 'PromoInterval']
train.drop(drop_columns + ['Customers'], axis=1, inplace=True)
test.drop(drop_columns, axis=1, inplace=True)
```

Finally, since we are sure that a closed shop is going to do 0 sales, we can strip the corresponding rows and remove the `Open` column. This way, the model will be able to focus only on real sales, and we will be able to standardize the Sales input properly.

```python
train.drop(train[train['Open'] == 0].index, inplace=True)
train.drop(['Open'], axis=1, inplace=True)
test.drop(test[test['Open'] == 0].index, inplace=True)
test.drop(['Open'], axis=1, inplace=True)
test.insert(loc=2, column='Sales', value=0)
```

```python
train.to_csv('data/processed/train.csv', index=False)
test.to_csv('data/processed/test.csv', index=False)
```
